```python
df_av_top = 0
df_av_top = df2[['Month','Year']].drop_duplicates().copy()
df_av_top.reset_index(inplace=True)

df_av_top['Day'] = 1
df_av_top['Datetime'] = pd.to_datetime(df_av_top[['Year','Month','Day']])


top10 = df2[df2['Rank']<6].copy()
llista_top5 = list(top10['Game'].unique())

for joc in llista_top5:
    df_av_top[joc] = 0
    aux = list(df2[df2['Game'] == joc]['Avg_viewers'].copy())
    for i in range(len(aux)):
        df_av_top[joc].iloc[70-i-1] = aux[-i-1]
```

```python
topor = df2.groupby(['Game','Year'])['Avg_viewers'].mean().reset_index(name='Avg_viewers')
topor = topor.sort_values(by=['Year', 'Avg_viewers'])

top10_any = pd.DataFrame(columns=list(topor.columns))

for any in set(topor['Year']):
    top10_any = top10_any.append(topor[topor['Year'] == any].tail(10), ignore_index=True)
    
topor2 = df.groupby('year')['Avg_viewers'].mean().reset_index(name='Avg_viewers')
topor2.rename(columns = {'year':'Year','Avg_viewers':'Avg_viewers', 'Game':'Game'}, inplace = True)
topor2['Game'] = 'Resta'

for any in set(topor['Year']):
    topor2['Avg_viewers'].iloc[any - 2016] -= sum(top10_any[top10_any['Year']==any]['Avg_viewers'])
    top10_any = top10_any.append(topor2[topor2['Year'] == any], ignore_index=True)

top10_any =  top10_any.round(0)
top10_any['Avg_viewers'] = top10_any['Avg_viewers'].astype(int)
```
